```python
import matplotlib.pyplot as plt
plt.scatter(df["Unemployment_Rate"], df["Stock_Index_Price"], color="blue")
plt.xlabel('Unemployment Rate', fontsize=14)
plt.ylabel('Stock Index', fontsize=14)
plt.title('Unemplyment vs Stock index plot')
plt.grid()
plt.show()
```

```python
plt.scatter(df["Interest_Rate"], df["Stock_Index_Price"], color="red")
plt.xlabel('Interest Rate', fontsize=14)
plt.ylabel('Stock Index', fontsize=14)
plt.title('Interestt vs Stock index plot')
plt.grid()
plt.show()
```

```python
x=df[['Interest_Rate', 'Unemployment_Rate']]
y=df['Stock_Index_Price']
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x,y)
```

```python
print('Intercept:n', model.intercept_)
print('Coefficients:n', model.coef_)
```

```python
model.predict([[2.75, 5.3]])
```

```python
import statsmodels.api as sm
x=sm.add_constant(x)
model=sm.OLS(y,x).fit()

predictions=model.predict(x)
print_model= model.summary()
print(print_model)
```

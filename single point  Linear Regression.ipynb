```python
import pandas as pd
import numpy as np
import math
from sklearn import linear_model
```

```python
df=pd.read_csv('C:/Users/AL MADINA COMPUTER/Desktop/Machine Learning/Lineardata.csv')
df.head(2)
```

```python
ar=df.drop('price', axis='columns')
```

```python
pr=df.price
```

```python
reg=linear_model.LinearRegression()
reg.fit(ar,pr)
```

```python
reg.predict([[3300]])
```

```python
d=pd.read_csv('C:/Users/AL MADINA COMPUTER/Desktop/Machine Learning/linearpredicted.csv')
```

```python
d['price']=reg.predict(d)
```

```python
d.to_csv('C:/Users/AL MADINA COMPUTER/Desktop/Machine Learning/linearpredicted1.csv', index=False)
```
